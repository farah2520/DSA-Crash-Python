```python

from abc import ABC, abstractmethod

class Vehical(ABC):
  @abstractmethod
  def go(self):
    pass

  @abstractmethod
  def stop(self):
    pass

class Car(Vehical):
  def go(self):
    print("you drive")

  def stop(self):
    print("you stop")

class Motorcycle(Vehical):
  def go(self):
    print("you ride the motor")

  def stop(self):
    print("you stop the motor")         # IF WE DONT PUT STOP METHOD HERE WE GET ERROR COZ PARENT CLASS HAS STOP METHOD

car=Car()
car.go()

mot= Motorcycle()
mot.go()

```

```python
class Shape:
  def __init__(self, color, filled):
    self.color= color
    self.filled= filled


# *************** in shape class we have a method describe
  def describe(self):
    print(f"it is {self.color} and {'filled' if self.filled else 'not filled'}")

class Circle(Shape):
  def __init__(self, color, filled, radius):
    # *********** to use the color and filled from parent class
    super().__init__(color, filled)
    self.radius= radius

  def describe(self):
    print(f"it has area of {3.14 * self.radius* self.radius}")

    # ***************** also using the describe method prom parent
    super().describe()

class Square(Shape):
  def __init__(self, color, filled, width):
    super().__init__(color, filled)
    self.width= width

class Triangle(Shape):
  def __init__(self, color, filled, width, height):
    super().__init__(color, filled)
    self.base= width
    self.height= height

cir= Circle('red', True, 5)
sq= Square('blue', False, 9)
print(cir.color)

print(sq.width)
cir.describe()       # if there was also a describe method in circle class, it willl use the childs version not the parents

# ***** Method overrriding

```

```python
# Polymorphism

class Shape:
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    self.radius= radius

  def area(self):
    return 3.14 * self.radius * self.radius

class Square(Shape):
  def __init__(self, side):
    self.side= side

  def area(self):
    return self.side * self.side

class Triangle(Shape):
  def __init__(self, base, height):
    self.base= base
    self.height= height

  def area(self):
    return 0.5 * self.base * self.height

# ////// pizza is unrelated to shape but we want to find the area of pizza
class Pizza(Circle):
  def __init__(self, toppings, radius):
    super().__init__(radius)
    self.toppings= toppings

shapes=[Circle(4), Square(5), Triangle(6,7), Pizza('onion',15)]   ## pizza area also calculated

for shape in shapes:
  print(f"{shape.area()}cm^2")


```

```python
## Aggregation

class Library:
  def __init__(self, name):
    self.name=name
    self.books= []

  def add_book(self, book):
    self.books.append(book)

  def list_books(self):
    return [f"{book.title} by {book.author} " for book in self.books]

class Book:
  def __init__(self, title, author):
    self.title = title
    self.author = author

library= Library('new york library')

book1= Book('hp', 'jk')
book2= Book('hobbit', 'ror')
book3= Book('puppy', 'bob')

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

print(library.name)
print(library.list_books())





```

```python
# static metho

class Employee:
  def __init__(self, name, position):
    self.name= name
    self.position= position

  # *************** instance method ***************
  def get_info(self):
    return f"{self.name}= {self.position}"

  # ****************** static method *****************

  @staticmethod

  def is_valid_position(position):
    valid_position= ['Manager','Cook', 'Cashier']
    return position in valid_position

employee1= Employee('simsim','Manager')
employee2= Employee('kitty','Cookr')
employee3= Employee('leo','Manager')

# ************ when caling static method, use classname not object
print(Employee.is_valid_position('rj'))

# ************ calling instance method
print(employee1.get_info())

```
