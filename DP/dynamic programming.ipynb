Fibonacci with **Recursion**


```python
# T= O(2^n) // no of noes* work done in each node

# overlapping subproblem
# optimal substructure- optimal sol of the whole problem can be found by optimal sol of sub problem

# Top-Down (Memoization)
# Bottom-Up (Tabulation) , Usually more space-efficient (can optimize space to O(1)/O(n)).

# *********************************************************************************************************************
# Fibonacci Numbers â€“ O(n) time, O(1) space

# Knapsack Problem (0/1, Unbounded, etc.)

# Longest Common Subsequence / Substring

# Edit Distance (Levenshtein Distance)

# Matrix Chain Multiplication

# Coin Change (Ways + Minimum Coins)

# Climbing Stairs / Minimum Cost Path

# DP on Grids (unique paths, min sum path)

# Partition Problems (Subset sum, Equal sum partition)

# DP on Trees/Graphs (Longest path, DP on DAGs, LIS)

# Top-Down (Memoization): recursive + cache results, solves only needed subproblems.
# Bottom-Up (Tabulation): iterative, fills entire table from base case.

def fib(n):
  if n<=1:
    return n
  else:
    return fib(n-1)+fib(n-2)

print(fib(8))
```

Fibonacci with **Memoisation**

```python
# use hashtable/ dictionary

def fib1(n,mem):
  if n not in mem:
    mem[n]= fib1(n-1,mem)+fib1(n-2,mem)
    return mem[n]
  else:
    return mem[n]


fib1(10,{0:0,1:1})
```

Min cost climbing **Recursive**

```python
def minCost(cost):
  n= len(cost)

  def helper(index):
    if index> n-1:
      return 0

    onestep= cost[index]+helper(index+1)
    twostep= cost[index]+helper(index+2)

    return min(onestep,twostep)

  return min(helper(0),helper(1))


```

```python
# with memoisation
def minCost(cost):
  n= len(cost)
  mincost= [-1]*(n)

  def helper(index):
    if index> n-1:
      return 0

    if mincost[index]!=-1:
      return mincost[index]

    onestep= cost[index]+helper(index+1)
    twostep= cost[index]+helper(index+2)

    return min(onestep,twostep)

  return min(helper(0),helper(1))
```

```python
# with memoization

def knapSack(W,wt,val,n):

  # dp_table
  dp= [[-1]*(W+1) for _ in range(n)]

  def helper(index,rem_weight):
    if index>n-1 or rem_weight==0:
      return 0

    if dp[index][rem_weight]!=-1:
      return dp[index][rem_weight]

    exclude= helper(index+1,rem_weight)
    include=0
    if wt[index]<=rem_weight:
      include= val[index]+helper(index+1,rem_weight-wt[index])

    dp[index][rem_weight]= max(exclude,include)
    return dp[index][rem_weight]

    return helper(0,W)

```

LCS **recursive**


```python
def longestCommonSubsequence(text1, text2):
  n=len(text1)
  m=len(text2)


  def helper(index1, index2):

    if index>n-1 or index2>m-1:
      return 0

    if text1[index1]==text2[index2]:
      return 1+helper(index1+1,index2+1)
    else:
      return max(helper(index1+1,index2+1),helper(index+1,index2+1))

    return helper(0,0)



```

```python
# LCS memoization

def longestCommonSubsequence(text1, text2):
  n=len(text1)
  m=len(text2)
  dp= [[-1]*(m) for _ in range(n)]


  def helper(index1, index2):

    if index>n-1 or index2>m-1:
      return 0

    if dp[index1][index2]!=-1:
      return dp[index1][index2]

    if text1[index1]==text2[index2]:
      dp[index1][index2]= 1+helper(index1+1,index2+1)

    dp[index1][index2]= max(helper(index1+1,index2+1),helper(index+1,index2+1))

    return helper(0,0)





```
