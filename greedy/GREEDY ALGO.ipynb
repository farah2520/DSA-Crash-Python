```python

# fractional knapsack- time - o(nlogn)
# space- o(1)
```

```python

def fractional_knapsack(W,arr,n):
  # sort arr by ratio profit/weight

  arr.sort(reverse= True,key= lambda x: x[0]/x[1])
  # [[1,2] ] 1
  remaining_weight= W
  value= 0
  n= len(arr)
  for i in range(n):
    if remaining_weight==0:
      break
    weight= min(remaining_weight, arr[i][1])
    value+= weight*arr[i][0]/arr[i][1]
    remaining_weight-= weight
  return value



```
